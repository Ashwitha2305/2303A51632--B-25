```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)
students = [f"Student_{i}" for i in range(1, 51)]
subjects = ["Math", "Science", "English"]
data = {
    "Student": np.random.choice(students, 150),
    "Subject": np.random.choice(subjects, 150),
    "Score": np.random.normal(loc=75, scale=10, size=150),  # mean=75, std=10
    "Gender": np.random.choice(["Male", "Female"], 150)
}
df = pd.DataFrame(data)
sns.set_theme(style="whitegrid", palette="muted")

plot = sns.catplot(
    data=df,
    x="Subject",
    y="Score",
    hue="Gender",
    kind="box",
    height=6,
    aspect=1.2,
    palette="Set2"
)
plot.fig.suptitle("Exam Results by Subject and Gender", fontsize=16, fontweight="bold")
plot.set_axis_labels("Subject", "Score (%)")
plot.add_legend(title="Gender")
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing(as_frame=True)
df = housing.frame

df.rename(columns={"MedHouseVal": "MedianHouseValue"}, inplace=True)

corr_matrix = df.corr()

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    linewidths=0.5,
    cbar_kws={'shrink': 0.8, 'label': 'Correlation Coefficient'}
)

plt.title("Correlation Matrix of California Housing Features", fontsize=16, fontweight="bold", pad=15)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
