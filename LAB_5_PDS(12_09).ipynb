```python
import pandas as pd
# Step 1: Create Customers Dataset
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5],
    'name': ['Ravi', 'Sneha', 'Amit', 'Priya', 'Kiran'],
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai']
})

print("Customers Dataset:")
print(customers, "\n")

# Step 2: Create Transactions Dataset

transactions = pd.DataFrame({
    'transaction_id': [101, 102, 103, 104, 105, 106, 107],
    'customer_id': [1, 2, 1, 3, 4, 5, 2],
    'amount': [200, 500, 150, 300, 700, 100, 400]
})

print("Transactions Dataset:")
print(transactions, "\n")

# Step 3: Merge Customers and Transactions

merged = pd.merge(transactions, customers, on='customer_id', how='inner')
print("Merged Dataset:")
print(merged, "\n")

# Step 4: Calculate Total Spend per Customer

customer_spend = merged.groupby(['customer_id', 'name'])['amount'].sum().reset_index()
customer_spend = customer_spend.rename(columns={'amount': 'total_spend'})
print("Total Spend per Customer:")
print(customer_spend, "\n")

# Step 5: Group by City to Analyze Spending

city_spend = merged.groupby('city')['amount'].sum().reset_index()
city_spend = city_spend.rename(columns={'amount': 'city_total_spend'})
print("City-wise Spending:")
print(city_spend, "\n")

# Step 6: Identify Top Spender

top_spender = customer_spend.loc[customer_spend['total_spend'].idxmax()]
print("Top Spender:")
print(top_spender, "\n")

print("Insights:")
print(f"The top spender is {top_spender['name']} with total spend {top_spender['total_spend']}.")
print(f"City-wise spending shows that {city_spend.loc[city_spend['city_total_spend'].idxmax(), 'city']} has the highest spend.")
```

```python
import pandas as pd

# Step 1: Create Patients Dataset

patients = pd.DataFrame({
    'patient_id': [1, 2, 3, 4, 5],
    'name': ['Ravi', 'Sneha', 'Amit', 'Priya', 'Kiran'],
    'age': [30, 45, 28, 35, 50],
    'gender': ['M', 'F', 'M', 'F', 'M']
})

print("Patients Dataset:")
print(patients, "\n")

# Step 2: Create Appointments Dataset

appointments = pd.DataFrame({
    'appointment_id': [101, 102, 103, 104, 105, 106, 107],
    'patient_id': [1, 2, 1, 3, 4, 5, 2],
    'doctor': ['Dr. Sharma', 'Dr. Gupta', 'Dr. Sharma',
               'Dr. Gupta', 'Dr. Rao', 'Dr. Rao', 'Dr. Gupta'],
    'date': ['2025-09-01', '2025-09-01', '2025-09-05',
             '2025-09-06', '2025-09-07', '2025-09-08', '2025-09-09']
})

print("Appointments Dataset:")
print(appointments, "\n")

# Step 3: Merge Patients and Appointments

merged = pd.merge(appointments, patients, on='patient_id', how='inner')
print("Merged Dataset:")
print(merged, "\n")

# Step 4: Count Appointments per Patient

patient_visits = merged.groupby(['patient_id', 'name'])['appointment_id'].count().reset_index()
patient_visits = patient_visits.rename(columns={'appointment_id': 'total_visits'})
print("Appointments per Patient:")
print(patient_visits, "\n")
# Step 5: Group by Doctor to Compute Workload

doctor_workload = merged.groupby('doctor')['appointment_id'].count().reset_index()
doctor_workload = doctor_workload.rename(columns={'appointment_id': 'total_appointments'})
print("Doctor Workload:")
print(doctor_workload, "\n")

# Step 6: Find Patient with Maximum Visits

top_patient = patient_visits.loc[patient_visits['total_visits'].idxmax()]
print("Patient with Maximum Visits:")
print(top_patient, "\n")
# Step 7: Insights

print(" Insights:")
print(f"The patient with the most hospital visits is {top_patient['name']} ({top_patient['total_visits']} visits).")
print(f"The busiest doctor is {doctor_workload.loc[doctor_workload['total_appointments'].idxmax(), 'doctor']} "
      f"with {doctor_workload['total_appointments']} appointments.")
```
